```python
import xarray as xr
import numpy as np
from pydap.client import open_url
from pydap.cas.urs import setup_session 
from datetime import datetime
```

```python
# NASA URS Authentication
# ## Set up a .netrc file from command line:
# >> echo "machine urs.earthdata.nasa.gov login UNAME password PWD" > ~/.netrc ; > ~/.urs_cookies
# >> chmod  0600 ~/.netrc
#
import netrc, fsspec, aiohttp
(username, account, password) = netrc.netrc().authenticators("urs.earthdata.nasa.gov")
fsspec.config.conf['https'] = dict(client_kwargs={'auth': aiohttp.BasicAuth(username, password)})
```

```python
def generate_paths(pattern,years,days):
    # Generate list of paths given url pattern and integer, or numpy array (e.g. np.arange) of years and days to load  
    from calendar import isleap

    day,year=np.meshgrid(days,years)
    day     = day .flatten()
    year    = year.flatten()
    
    paths_list = [None]*len(day)
    for ii in np.arange(len(day)):
        if isleap(year[ii]):
            maxday=366
        else:
            maxday=365
        paths_list[ii] = pattern.format(year[ii],day[ii],np.min((maxday,day[ii]+np.mean(np.diff(days))-1)))
    return paths_list

def nasa_ocean_color_preprocess(ds):
    # preprocess each file to add time dimension, allowing merge
    ds=ds.assign_coords({"time": datetime.strptime(ds.time_coverage_start, '%Y-%m-%dT%H:%M:%S.%fZ')})
    return ds

def read_nasa_opendap(paths, dim, transform_func=None):
    # Read data files at path location using pydap
    # This routine is hampered by timeouts...not sure why.
    def process_one_path(path):
        # use a context manager, to ensure the file gets closed after use
        with xr.open_dataset(xr.backends.PydapDataStore(open_url(path,timeout=360))) as ds:
            print(path)
            # transform_func should do some sort of selection or
            # aggregation
            if transform_func is not None:
                ds = transform_func(ds)
            # load all data from the transformed dataset, to ensure we can
            # use it after closing each original file
            ds.load()
            return ds
    datasets = [process_one_path(p) for p in paths]
    combined = xr.concat(datasets, dim)
    return combined

def read_nasa_intake(paths, dim, transform_func=None):
    # Read data files at path location using intake
    # This routine does not seem to time out
    import intake
    import intake_xarray
    def process_one_path(path):
        # use a context manager, to ensure the file gets closed after use
        with intake.open_opendap(urlpath=path,engine='netcdf4').to_dask() as ds:
            print(path)
            # transform_func should do some sort of selection or
            # aggregation
            if transform_func is not None:
                ds = transform_func(ds)
            # load all data from the transformed dataset, to ensure we can
            # use it after closing each original file
            ds.load()
            return ds
    datasets = [process_one_path(p) for p in paths]
    combined = xr.concat(datasets, dim)
    return combined

url="https://oceandata.sci.gsfc.nasa.gov/opendap/hyrax/MODISA/L3SMI/{0}/{1:03d}/A{0}{1:03d}{0}{2:03.0f}.L3m_8D_CHL_chlor_a_9km.nc?lon,lat,chlor_a"
ds3 = read_nasa_intake(generate_paths(url,np.arange(2003,2014),np.arange(1,367,8)), dim='time',
                           transform_func=lambda ds: nasa_ocean_color_preprocess(ds))
# Remove protected attribute (https://github.com/pydata/xarray/issues/2822)
del ds3.attrs['_NCProperties']
```

```python
# Rechunk data (6.4Mb chunks, could be larger, but fits nicely)
ds3=ds3.chunk({'lat':270,'lon':270,'time':23})
ds3
```

```python
# Write out to netcdf...could take a while
ds3.to_netcdf(path='/Users/jonyo/Downloads/modis-chla-8d-2003-2013.nc',mode='w')
```
